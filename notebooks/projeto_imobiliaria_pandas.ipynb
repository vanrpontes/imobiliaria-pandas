# 2º Análise Exploratória de Dados (EDA)
## 1º Qual o Valor Médio de Aluguel por Tipo de Imóvel?

```python
dados.head()
```

```python
dados['Valor'].mean()
```

```python
# O Método .groupby() agrupa os valores por um parâmetro específico
# A chamada .groupby("nome_da_coluna") divide um DataFrame em grupos, aplica uma função a cada grupo e combina os resultados.

dados.groupby('Tipo').mean(numeric_only=True)
```

```python
# Um colchete [] retorna apenas uma Séries; dois colchetes [[]] retornam um DataFrame.
# .groupby() divide um DataFrame em grupos, aplica uma função a cada grupo e combina os resultados.

dados.groupby('Tipo')['Valor'].mean()
```

```python
# Usamos dois colchetes [[]] para selecionar uma ou mais colunas como DataFrame.
# Um colchete [] retorna apenas uma Séries; dois colchetes [[]] retornam um DataFrame.
# Isso é útil quando queremos manter o formato de tabela para operações como sort_values(), agregações com múltiplas colunas, ou para encadear métodos de forma consistente.

dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_media_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_media_preco_tipo.plot(kind='barh', figsize=(14, 10), color ='blue', xlabel='Média Aluguel', ylabel='Tipo de Imóvel')
```

#### Remover os imóveis de Tipo Comercial da base de dados


```python
# Para consultar os valores únicos de uma coluna usa-se o método unique()

dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                     'Prédio Inteiro', 'Loja/Salão',
                     'Galpão/Depósito/Armazém',
                     'Casa Comercial', 'Terreno Padrão',
                     'Loja Shopping/ Ct Comercial',
                     'Box/Garagem', 'Chácara',
                     'Loteamento/Condomínio', 'Sítio',
                     'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
# Para filtrar linhas usando expressões semelhantes às de SQL, mas escritas como strings, usamos o método query do Pandas.

dados.query("@imoveis_comerciais in Tipo")
```

```python
# Agora para selecionar o exato oposto da seleção anterior, basta acrescentar o operador 'not' antes do 'in'

dados.query("@imoveis_comerciais not in Tipo")
```

```python
# Salvando a seleção 'não em imoveis_comercial' em uma nova variável

df = dados.query("@imoveis_comerciais not in Tipo")

df
```

```python
# Para verificar se agora temos apenas os Tipos que iremos analisar (não comerciais), usamos o método unique de novo.

df.Tipo.unique()
```

```python
df_preco_medio_tipo = df.groupby('Tipo')['Valor'].mean().sort_values().round(2)

df_preco_medio_tipo.plot(kind='barh', figsize=(14, 10), color='purple')

df_preco_medio_tipo
```

# 3º Tratamento
## Verificar e Tratar Valores Nulos

```python
# Vamos descobrir quais colunas da base tem valores nulos com o .isnull()

df.isnull()
```

```python
# Para melhor visualização, podemos somar os valores nulos

df.isnull().sum()
```

```python
# A função fillna() substitui todos os valores nulos por um parâmetro dentro dos parênteses, neste caso o '0'.

df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
# Vamos consultar se há valores zero nas colunas 'Valor' ou 'Condominio'

df.query('Valor == 0 | Condominio ==0')
```

```python
df.query('Valor == 0 or Condominio == 0').index
```

```python
registros_a_remover = df.query('Valor == 0 or Condominio == 0').index
```

```python
# Método drop() para remover as linhas com valor ou condominio igual a 0
# parâmetros axis=0 (linhas) - o 1 são para colunas
# parâmetro inplace=True remove e guarda esse tratamento sem a necessidade de gaurdar isso em uma variável

df.drop(registros_a_remover, axis=0, inplace=True)
```

```python
df.query('Valor == 0 or Condominio == 0')
```
